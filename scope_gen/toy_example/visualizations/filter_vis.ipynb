```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

set_seed(3)

# Generate random data from a standard Gaussian distribution
data = np.random.normal(loc=0, scale=1, size=(10, 2))

# Define the mean and covariance for a standard Gaussian distribution
mean = np.array([0, 0])
covariance = np.array([[1, 0], [0, 1]])

# Create a grid of points where the density will be evaluated
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
positions = np.dstack((X, Y))

# Evaluate the density on the grid using the theoretical Gaussian formula
density = multivariate_normal(mean, covariance)
Z = density.pdf(positions)

# Plotting
fig, ax = plt.subplots()
contour = ax.contour(X, Y, Z, colors='red', levels=6, zorder=1)  # adjust levels for more/less contours
ax.clabel(contour, inline=True, fontsize=8)
ax.scatter(data[:, 0], data[:, 1], alpha=0.6, color='blue')
ax.set_xlabel('$y_1$')
ax.set_ylabel('$y_2$')
ax.legend()
plt.show()
```
